```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
plt.plot(x, s)
plt.title('Sine X')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.scatter(x, s)
plt.title('Sine X')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.plot(x, c)
plt.title('Cos X')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.scatter(x, c)
plt.title('Cos X')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.plot(x, t)
plt.title('Tangent X')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.scatter(x, t)
plt.title('Tangent X')
plt.xlabel('X')
plt.ylabel('Y')
```
