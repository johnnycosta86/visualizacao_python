```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)

a_s = np.arcsin(x)
a_c = np.arccos(x)
a_t = np.arctan(x)

x2 = np.exp(x)
```

Object Oriented way - Sub Plot

```python
fig, aux = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

aux[0].plot(x, s)
aux[0].set_title('Sin')
aux[0].set_xlabel('X')
aux[0].set_ylabel('Y')

aux[1].plot(x, s)
aux[1].set_title('Sin')
aux[1].set_xlabel('X')
aux[1].set_ylabel('Y')
```

```python
fig, aux = plt.subplots(nrows=1, ncols=6, figsize=(10, 5))

aux[0].plot(x, s)
aux[0].set_title('Sin')
aux[0].set_xlabel('X')
aux[0].set_ylabel('Y')

aux[1].plot(x, c)
aux[1].set_title('Cos')
aux[1].set_xlabel('X')
aux[1].set_ylabel('Y')

aux[2].plot(x, t)
aux[2].set_title('Tan')
aux[2].set_xlabel('X')
aux[2].set_ylabel('Y')

aux[3].plot(x, a_s)
aux[3].set_title('Arc Sin')
aux[3].set_xlabel('X')
aux[3].set_ylabel('Y')

aux[4].plot(x, a_c)
aux[4].set_title('Arc Cos')
aux[4].set_xlabel('X')
aux[4].set_ylabel('Y')

aux[5].plot(x, a_t)
aux[5].set_title('Arc Tan')
aux[5].set_xlabel('X')
aux[5].set_ylabel('Y')

plt.tight_layout()
```

```python
fig, aux = plt.subplots(nrows=3, ncols=3, figsize=(10, 5))

aux[1][2].plot(x, x2)
aux[1][2].set_title('X^2')
aux[1][2].set_xlabel('X')
aux[1][2].set_ylabel('Y')

plt.tight_layout()
```
