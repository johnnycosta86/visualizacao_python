```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10, 5))
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.plot(x, s, color='r')
aux.plot(x, s+1, color='g')
aux.plot(x, s+2, color='c')
aux.plot(x, s+3, color='y')
aux.plot(x, s+4, color='b', marker='o')
aux.plot(x, s+5, color='r', marker='o', markersize=10)
aux.plot(x, s+6, color='g', marker='o', markersize=5, lw=0.5)
aux.plot(x, s+7, color='b', marker='o', markersize=5, lw=3.5)
aux.plot(x, s+7, color='c', marker='d', markersize=5, lw=3.5)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.plot(x,s, color='m', marker="p", markersize=10, markerfacecolor='b', linestyle='dashed')
```
