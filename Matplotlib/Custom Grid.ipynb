```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.grid(color='b')
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.grid(color='b', linestyle='dashed')
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.grid(color='b', linestyle='dashdot')
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.grid(color='b', linestyle='dotted')
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.grid(color='b', linestyle='solid')
ax.plot(x, x*x)
```
