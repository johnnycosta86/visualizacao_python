```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
plt.plot(x, s, x, c)
```

```python
plt.plot(x, s, x, c, x, t)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, s, 'b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('1st Chart')

plt.subplot(1, 2, 2)
plt.plot(s, x, 'r')
plt.xlabel('Y')
plt.ylabel('X')
plt.title('2st Chart')

# Add x symbol between graphs
plt.tight_layout() 
```
