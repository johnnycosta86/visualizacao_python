```python
import matplotlib.pyplot as plt
```

```python
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]
z1 = [200, 25, 400, 330, 100]
```

```python
title: str = 'Scatterplot Graphs Title'
xlabel: str = 'Label X'
ylabel: str = 'Label Y'
```

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.plot(x1, y1, color='b', linestyle='--')
plt.scatter(x1, y1, label='My points', color='r')
plt.legend()
plt.savefig('figure01.png')
```
