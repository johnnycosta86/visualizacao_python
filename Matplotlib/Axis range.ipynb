```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.plot(x, s, color='b', label='Sin')
aux.legend()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.set_xlim([-5, 20]) # Allows other numbers in X
aux.plot(x, s, color='b', label='Sin')
aux.legend()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.set_xlim([-5, 20]) # Allows other numbers in X
aux.set_ylim([-2, 2]) # Allows other numbers in Y
aux.plot(x, s, color='b', label='Sin')
aux.legend()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.set_xlim([-5, 0.5]) # Allows other numbers in X
aux.set_ylim([-1, 1.5]) # Allows other numbers in Y
aux.plot(x, s, color='r', label='Sin')
aux.legend()
```
