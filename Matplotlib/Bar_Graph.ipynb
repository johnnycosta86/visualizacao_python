```python
import matplotlib.pyplot as plt
```

```python
data = open('brazil_population.csv').readlines()
```

```python
x = []
y = []

for i in range(len(data)):
    if i != 0:
        line = data[i].split(';')
        x.append(int(line[0]))
        y.append(int(line[1]))
```

```python
title: str = 'Growth of the Brazilian Population 1980-2016'
xlabel: str = 'Year'
ylabel: str = 'Population'
```

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.plot(x, y)
plt.legend()
```

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.bar(x, y)
plt.legend()
```

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.bar(x, y) # Bar graph
plt.plot(x, y) # Line graph
plt.legend()
```

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.bar(x, y, color='#e4e4e4') # Bar graph
plt.plot(x, y) # Line graph
plt.legend()
```

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.bar(x, y, color='#e4e4e4') # Bar graph
plt.plot(x, y, color='k', linestyle='--') # Line graph
plt.legend()
```

```python
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.bar(x, y, color='#e4e4e4') # Bar graph
plt.plot(x, y, color='k', linestyle='--') # Line graph
# plt.scatter(x, y, color='k',) # Point graph
plt.legend()
plt.savefig('brazil_population.png', dpi=300)
```
