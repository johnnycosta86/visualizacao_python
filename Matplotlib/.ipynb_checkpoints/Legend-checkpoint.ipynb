```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10, 5))
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.plot(x, s, color='r', label='Sin')
aux.plot(x, c, color='g', label='Cos')
aux.legend()
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10, 5))
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.plot(x, s, color='r', label='Sin')
aux.plot(x, c, color='g', label='Cos')
aux.legend(loc=7)
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(6, 4))
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.plot(x, s, color='r', label='Sin')
aux.plot(s, x, color='b', label='Inverse of Sin')
aux.legend(loc=7) # loc change box legend position [1 to 6]
```
