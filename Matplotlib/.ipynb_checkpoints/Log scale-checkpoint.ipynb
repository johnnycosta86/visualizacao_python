```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.plot(x, x*x*x*x*x, color='b', label='Sin')
aux.legend()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.set_yscale('log')  # print in log scale
aux.plot(x, x*x*x*x*x, color='b', label='Sin')
aux.legend()
```
