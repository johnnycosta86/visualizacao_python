```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
x2 = np.exp(x)
```

```python
plt.plot(x, s, x, c)
```

```python
fig = plt.figure() # Increase image size
ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax2 = fig.add_axes([0.4, 0.65, 0.25, 0.25])

ax1.plot(x, s, color='m')
ax2.plot(x, s, color='g')

ax1.set_title('Sin Greater')
ax2.set_title('Sin Min')
```

```python
a = np.arange(0,10,0.2)
b = a ** 2
 
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax2 = fig.add_axes([0.15, 0.3, 0.25, 0.25])

ax1.plot(a,b, color='magenta')
ax2.plot(a,b, color='green')
 
ax1.set_title('Bigger graph')
ax2.set_title('Smaller graph')
```
