```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
for style in plt.style.available:
    plt.style.use(style)
    fig = plt.figure(figsize=(18, 5))
    aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
    aux.set_title(style)
    aux.plot(x, s)
```
