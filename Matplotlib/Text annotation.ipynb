```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.text(4, 60, 'y = X Squared')
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.text(4, 60, '$y = x^2$')
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.text(4, 60, '$y = x^2$', fontsize=15)
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.text(4, 60, '$y = x^2$', fontsize=15, color='b')
ax.plot(x, x*x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.text(1, 60, '$y = x^2$', fontsize=15, color='b')
ax.plot(x, x*x)
```
