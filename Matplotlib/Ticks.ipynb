```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array from 0 to 10 with 0.2 from each value
x = np.arange(0, 10, 0.2)
s = np.sin(x)
c = np.cos(x)
t = np.tan(x)
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.plot(x, s, color='b', label='Sin')
aux.legend()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.set_xticks([1,3,5,7,9]) # Define default values in x
aux.plot(x, s, color='b', label='Sin')
aux.legend()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.set_xticks([1,3,5,7,9]) # Define default values in x
aux.set_yticks([-1,-0.5,0,0.5,1]) # Define default values in y
aux.plot(x, s, color='b', label='Sin')
aux.legend()
```

```python
x = np.arange(0, 10, 0.5)

plt.style.use('ggplot')
fig = plt.figure()
aux = fig.add_axes([0.1, 0.1, 0.9, 0.9])
aux.set_xticks([1,2.5,2.5,5.0, 6.0,8.0, 10.0]) # Define default values in x
aux.set_yticks([0,20,40,60,80,100]) # Define default values in y
aux.plot(x, x*x, color='r', label='Sin')
aux.legend()
```
