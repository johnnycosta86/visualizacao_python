```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import plot, arange, xlabel, ylabel, title, grid, axhline, axvline, subplots, axis
```

```python
x = arange(0,4,0.5)
y = 3 * x
```

```python
plot(x, y, '^')
xlabel('Variable X')
ylabel('Function y = 3 * x')
title('Graph name')
grid(True)
```

```python
plot(x, y, '^')
xlabel('Variable X')
ylabel('Function y = 3 * x')
title('Graph name')
grid(True)
axhline(y=0, color='k')
axvline(x=0, color='k')
```

```python
fig, ax = subplots()
ax.plot(x, y)
ax.grid(True, which='both')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.yaxis.tick_left()
ax.xaxis.tick_bottom()

xlabel('Variable X', horizontalalignment='right', position=(1, 10))
ylabel('Function y = 3 * x', verticalalignment='center', position=(1, 0.7))
title('Graph name')

```

```python
plot(x, y, 'v')
xlabel('Variable X')
ylabel('Function y = 3 * x')
title('Graph name')
grid(True)
axis([-4, 3, -7, 7])
```
