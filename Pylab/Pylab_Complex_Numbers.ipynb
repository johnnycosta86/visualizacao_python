```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import polar, angle, plot, text, axis, grid, axhline, axvline
```

```python
z1 = 2 - 5j
z2 = -3 + 2j
```

```python
polar([0, angle(z1)], [0, abs(z1)], marker='o')
```

```python
polar([0, angle(z2)], [0, abs(z2)], marker='*')
```

```python
plot([0, z1.real], [0, z1.imag], 'ro-')
text(z1.real + 0.2, z1.imag + 0.2, 'z1')
plot([0, z2.real], [0, z2.imag], 'b*--')
text(z2.real + 0.2, z2.imag + 0.2, 'z2')
axis([-4, 3, -7, 7])
axhline(y=0, color='k')
axvline(x=0, color='k')
grid(True)
```
