```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import plot, arange, xlabel, ylabel, grid, yscale, xscale, subplot, title, subplots_adjust, axis
from math import sin, pi, exp
```

```python
def polynomial(x):
    return x**2 - 2*x - 3
```

```python
def trigonometry(x):
    return 2 * sin(x/2 + pi/3)
```

```python
def exponential(x):
    return 2 * exp(-x/4)
```

```python
def logarithm(x):
    return 3 * np.log10(x**2)
```

```python
x = arange(-5.0, 7.0, 0.01)
y = polynomial(x)
plot(x, y)
xlabel('x')
ylabel('y = f(x)')
title('Quadratic Function y = x**2 - 2*x - 3')
grid(True)
```

```python
x = np.arange(0, 6*pi + pi/3, 0.01)
y = [trigonometry(v) for v in x]
plot(x, y)
xlabel('x')
ylabel('y = f(x)')
title('Trigonometry Function y = 2sin(x/2 + pi/3)')
grid(True)
axis([0, 20, -2.5, 2.5])
```

```python
x = np.arange(-1, 9, 0.01)
y = [exponential(v) for v in x] 
plot(x, y)
xlabel('x')
ylabel('y = f(x)')
title('Exponential Function y = 2*exp(-x/4)')
grid(True)
```

```python
x = np.arange(-1, 9, 0.01)
y = [logarithm(v) for v in x] 
plot(x, y)
xlabel('x')
ylabel('y = f(x)')
title('Logarithm Function y = 3*log(x**2)')
yscale('log')
grid(True)
```
