```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import plot, arange, yscale, xscale, subplot, title, subplots_adjust
```

```python
x = arange(0,4,0.5)
y = 2 * x
y1 = 3 * y
y2 = y**2
```

```python
subplot(121)
plot(x, y, 'g^', x, y1, 'rd', x, y2, 'b*')
title('Graph Title')

subplot(122)
yscale('log')
plot(x, y, 'g^', x, y1, 'rd', x, y2, 'b*')
```

```python
subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=1.1, wspace=0.2, hspace=0.4)

subplot(221)
plot(x, y, 'g^', x, y1, 'ro', x, y2, 'bs')
title('Graph 1 Title')

subplot(222)
yscale('log')
plot(x, y, 'k*', x, y1, 'cd', x, y2, 'mp')
title('Graph 2 Log Y Title')

subplot(223)
plot(x, y/2, 'yh', x, 2*y1, 'wo', x, y2/3, 'gx')
title('Graph 3 Title')

subplot(224)
xscale('log')
plot(x**2, y, 'cv', x**3, y1, 'k+', x**2.5, y2/2, 'b.')
title('Graph 4 Log X Title')
```
